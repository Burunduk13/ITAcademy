<h3>Mostreu-me a veure que passa quan eliminem el canal G Verd o B Blau.

Mostreu-me a veure què passa quan eliminem el canal G Verd o B Blau. Hauries d'utilitzar la indexació per seleccionar el canal que voleu anul·lar.

Utilitzar el mètode, mpimg.imsave () de la llibreria importada, per guardar les imatges modificades i que haureu de pujar al vostre repositori a github.</h3>

```python
import matplotlib.image as mpimg
import matplotlib.pyplot as plt   
#estudieu el metodde mpimg.imread(()
img = mpimg.imread('casey-b.png')

imgGreen_0 = img.copy()
imgBlue_0 = img.copy()

imgGreen_0[:, :, 1] = 0
imgBlue_0[:, :, 2] = 0
mpimg.imsave("green_0.png", imgGreen_0)
mpimg.imsave("blue_0.png", imgBlue_0)
```

```python
fig, axes = plt.subplots(1, 3)

axes[0].imshow(img)
axes[0].set_title('Original')

axes[1].imshow(imgGreen_0)
axes[1].set_title('Green = 0')

axes[2].imshow(imgBlue_0)
axes[2].set_title('Blue = 0')

for ax in axes:
    ax.set_xticks([])
    ax.set_yticks([])

plt.show()
```
