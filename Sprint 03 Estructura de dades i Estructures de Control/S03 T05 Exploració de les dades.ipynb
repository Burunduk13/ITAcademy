Descarrega el data set Airlines Delay: Airline on-time statistics and delay causes i carrega’l a un pandas Dataframe. Explora les dades que conté, i queda’t únicament amb les columnes que consideris rellevants.

```python
import pandas as pd
import numpy as np
df=pd.read_csv ("data\DelayedFlights.csv")
df=df[df["AirTime"]>10]
dfA=df.loc[:,('UniqueCarrier','Origin', 'Dest','AirTime', 'ArrDelay','DepDelay','Distance','TaxiIn', 'TaxiOut', 'WeatherDelay', 'NASDelay', 'SecurityDelay', 'LateAircraftDelay')]
dfA
```

Fes un informe complet del data set:.

- Resumeix estadísticament les columnes d’interès
- Troba quantes dades faltants hi ha per columna
- Crea columnes noves (velocitat mitjana del vol, si ha arribat tard o no...)
- Taula de les aerolínies amb més endarreriments acumulats
- Quins són els vols més llargs? I els més endarrerits?

Etc.

```python
dfMean=dfA.groupby(["UniqueCarrier"]).agg({"AirTime" :np.mean, "ArrDelay":[min,np.mean,max],"DepDelay" :[min,np.mean,max], "TaxiIn" :np.mean,"TaxiOut":np.mean, 'WeatherDelay':np.mean, 'NASDelay':np.mean, 'SecurityDelay':np.mean, 'LateAircraftDelay':np.mean})
#.sort_values("ArrDelay",ascending=False).reset_index(drop=False)
dfMean
```

```python
Falta=dfA.isnull().sum()
print (Falta)
```

```python
#Crea columnes noves (velocitat mitjana del vol, si ha arribat tard o no...)
dfA["aceleration"]=dfA["Distance"]/dfA["AirTime"]
dfA["Delay"]=dfA["ArrDelay"].apply(lambda x: "late" if x>10 else "ontime")

dd=dfA.groupby(["UniqueCarrier","Origin","Dest","Delay"]).agg({"AirTime" :np.mean, "ArrDelay":[min,np.mean,max],"DepDelay" :[min,np.mean,max], "TaxiIn" :np.mean,"TaxiOut":np.mean, 'WeatherDelay':np.mean, 'NASDelay':np.mean, 'SecurityDelay':np.mean, 'LateAircraftDelay':np.mean})
dd
```

```python
#Taula de les aerolínies amb més endarreriments acumulats
dfLateSum=dfA.groupby("UniqueCarrier")[["ArrDelay"]].sum().sort_values("ArrDelay",ascending=False)
dfLateCount=dfA.groupby("UniqueCarrier")[["ArrDelay"]].count().sort_values("ArrDelay",ascending=False)
print(dfLateSum)
print("----------------------------------")
print(dfLateCount)
```

```python
vuelo=dfA.sort_values("AirTime",ascending=False)
#time
print ("El vuelo más largo es de ", vuelo.iloc[0,1], "a ", vuelo.iloc[0,2], " - time")
#distance
vueloD=dfA.sort_values("Distance",ascending=False)
print ("El vuelo más largo es de ", vueloD.iloc[0,1], "a ", vueloD.iloc[0,2], " - distance")
```
