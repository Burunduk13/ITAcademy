# SMOTEÂ¶
SMOTE (Syntetic Minority Oversampling Technique) is an oversampling technique where the synthetic samples are generated for the minority class.

```python

df_smote=df.copy()
df_smote["Medal"]=df_smote["Medal"].apply(lambda x: 1 if str(x) != "nan" else 0)
df_smote.head()
```

```python
df_smote['Medal'].value_counts()
```

```python
df_smote=df_smote[['Sex', 'Age', 'Height', 'Weight', 'Team', 'NOC','Year', 'Sport', 'Medal']]
df_smote.dropna(inplace=True)
df_smote.head()
```

```python
df_smote['Medal'].value_counts()
```

```python
df_smote.info()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score

X = df_smote.drop(["Medal"], axis=1)
y = df_smote["Medal"]
X.head()
```

```python
#MULTIconvert text data to numeric before applying SMOTE
from sklearn import preprocessing
class MultiColumnLabelEncoder:

    def __init__(self, columns=None):
        self.columns = columns # array of column names to encode


    def fit(self, Xc, y=None):
        self.encoders = {}
        columns = Xc.columns if self.columns is None else self.columns
        for col in columns:
            self.encoders[col] = preprocessing.LabelEncoder().fit(Xc[col])
        return self


    def transform(self, Xc):
        output = Xc.copy()
        columns = Xc.columns if self.columns is None else self.columns
        for col in columns:
            output[col] = self.encoders[col].transform(Xc[col])
        return output


    def fit_transform(self, Xc, y=None):
        return self.fit(X,y).transform(Xc)


    def inverse_transform(self, Xc):
        output = Xc.copy()
        columns = Xc.columns if self.columns is None else self.columns
        for col in columns:
            output[col] = self.encoders[col].inverse_transform(Xc[col])
        return output
```

```python
multi = MultiColumnLabelEncoder(columns=['Sex','Team','NOC','Sport'])

Xc = multi.fit_transform(X)
Xc.head()
```

```python
from imblearn.over_sampling import SMOTE
X_resampled, y_resampled = SMOTE().fit_resample(Xc, y)
y_resampled.value_counts()
```

```python
X_resampled.head()
```

```python
invX_resample = multi.inverse_transform(X_resampled)
invX_resample.head()
```

```python
df_new = invX_resample
df_new ["Medal"] = y_resampled
df_new.head()
```

# Exercises 3
It continues with the sports theme dataset and generates a sample using the Reservoir sampling method

```python
import random
SAMPLE_COUNT = 10000

# Force the value of the seed so the results are repeatable
random.seed(12345)

sample_index = []
for index, line in enumerate(df.index):
        # Generate the reservoir
        if index < SAMPLE_COUNT:
            sample_index.append(line)
        else:
                # Randomly replace elements in the reservoir
                # with a decreasing probability.
                # Choose an integer between 0 and index (inclusive)
            r = random.randint(0, index)
            if r < SAMPLE_COUNT:
                sample_index[r] = line
print(sample_index)
```

```python
df_reservoir=df[df.index.isin(sample_index)]
df_reservoir.head()
```
