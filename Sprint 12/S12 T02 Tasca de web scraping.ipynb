**Webscraping with BeautifulSoup**

```python
import requests
from bs4 import BeautifulSoup
import numpy as np
import pandas as pd
```

```python
URL = "https://www.bolsamadrid.es/esp/aspx/Mercados/Precios.aspx?indice=ESI100000000"
page = requests.get(URL)

soup = BeautifulSoup(page.content, "html.parser")
```

```python
soup
```

```python
#find a html table in the web page
table = soup.find_all('table',class_='TblPort')[1]
rows = table.find_all('tr')
table
```

```python
rows
```

```python
tab_data = [[cell.text for cell in row.find_all(['th', 'td'])] for row in rows]
df = pd.DataFrame(tab_data)
df.head()
```

```python
header = df.iloc[0]
df = df[1:]
df.columns = header
df.head()
```

**Webscraping with Selenium**

```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.common.by import By
```

```python
s = Service(r'C:\Users\Usuario\AppData\Local\Chrome\chromedriver.exe')
driver = webdriver.Chrome(service=s)
driver.get("https://www.bolsamadrid.es/esp/aspx/Mercados/Precios.aspx?indice=ESI100000000")
print(driver.title)
print(driver.current_url)
#driver.close()
```

```python
id="ctl00_Contenido_tblAcciones"
```

```python
element=driver.find_element(By.ID,id)
print(element)
```

```python
_rows=[]
cols=[]
table=element.find_elements_by_xpath(".//tr")
for col_name in table[0].find_elements_by_xpath(".//th"):
    cols.append(col_name.get_attribute('textContent'))


for row in table[1:]:
    df_cell=[]
    for cell in row.find_elements_by_xpath(".//td"):
        df_cell.append(cell.get_attribute('textContent'))
    _rows.append(df_cell)

df_selenium = pd.DataFrame(_rows)
df_selenium.columns = cols
df_selenium.head()
```

```python
#export to csv

df_selenium.to_csv('precios_sesion.csv', index=False)
```

* Nombre - name of the companies
* Últ. - last price

* Máx. - Maximum Price
* Mín. - Minimum Price
* Volumen - amount of the hiring of a certain degree in a period of time
* Efectivo (miles €) - Amount in euros. The traded cash of a transaction is the result of multiplying the number of shares traded by the price of each share (this is where the cash in euros is obtained; one of the measures used to measure the liquidity of a security)

* Hora - time the prices were checked

```python
import scrapy
from scrapy.crawler import CrawlerProcess
```

```python
# write items to a JSON file
import json

class JsonWriterPipeline(object):

    def open_spider(self, spider):
        self.file = open('imdb.jl', 'w')
        self.file1 = open('imdb.json', 'w')

    def close_spider(self, spider):
        self.file.close()
        self.file1.close()

    def process_item(self, item, spider):
        line = json.dumps(dict(item)) + "\n"
        self.file.write(line)
        self.file1.write(line)
        return item
```

```python
# Logging from Spiders
import logging

class IMDB(scrapy.Spider):
    name = "imdb"
    start_urls = ['https://www.imdb.com/chart/toptv/?ref_=nv_tvv_250']
    custom_settings = {
        'LOG_LEVEL': logging.WARNING,
        'ITEM_PIPELINES': {'__main__.JsonWriterPipeline': 1}, # Used for pipeline 1
        'FEED_FORMAT':'json',                                 # Used for pipeline 2
        'FEED_URI': 'imdb.json'                        # Used for pipeline 2
    }
    
    def parse(self, response):
        for db in response.css('tbody.lister-list tr'):
            yield {
                'position': db.css('td.titleColumn::text').extract_first(),
                'name': db.css('td.titleColumn a::text').extract_first(),
                'year': db.css('td.titleColumn span.secondaryInfo::text').extract_first(),
                'artists':db.css("td.titleColumn a::attr(title)").extract_first(),
                'rating':db.css('td strong::text').extract_first(),                
                'poster':db.css("td.posterColumn img::attr(src)").extract_first(),
                'moive_link':db.css("td.posterColumn a::attr(href)").extract_first(),
            }
```

```python
process = CrawlerProcess({
    'USER_AGENT': 'Mozilla/4.0 (compatible; MSIE 7.0; Windows NT 5.1)'
})

process.crawl(IMDB)
process.start()
```
